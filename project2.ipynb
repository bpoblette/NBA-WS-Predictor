Here are the imports needed to complete the analysis

```python
import pandas as pd
```

Import the data and clean it up

```python
player_stats = pd.read_csv('all_seasons.csv')
```

Here I put the shape and the dataset's headers. This section I just followed the youtube video and did the functions showed in the video.

```python
print("The number of (rows, columns) in the dataset are:", player_stats.shape)
print(player_stats.head)
print(player_stats.describe())
```

1. Discovering: Here we can see the datasets min, max, median for several features I found interesting. I chose field goal percentage, points, winshares, and true shooting percentage because eventually I think I'll try to make a model that predicts the WS based off the other 3 variables

```python
important_features = ['net_rating%', 'PTS', 'WS', 'TS%']  # Example of important features
for feature in important_features:
    print(f"Statistics for {feature}:")
    print("Min:", player_stats[feature].min())
    print("Max:", player_stats[feature].max())
    print("Median:", player_stats[feature].median())
    print("Mean:", player_stats[feature].mean())
    print()
```

2. Structuring. In this section I just wanted to play with the WS column. I made it so that if the WS is higher than 2 then the value would change to 1. If it's less that 2 then the value becomes 0. This could be a way of showing that a player is seen as impactful if their winshares are higher than 2.

```python
def convert_ws(value):
    if value > 2:
        return 1
    else:
        return 0

player_stats['WS_01'] = player_stats['WS'].apply(convert_ws)
print(player_stats[['WS', 'WS_01']].head())
```

3. Cleaning. For this I just changed the missing values for "WS/48" to the mean "WS/48". Minutes played wasn't a stat that used to be tracked and because of this we can't calculate a player's winshares per 48 minutes. I don't think it actually makes sense to fill it with the man in an actual analysis of the data but I am just using this as practice using the fill functions.

```python
avg_ws_48 = player_stats['WS/48'].mean()

player_stats['WS/48'].fillna(avg_ws_48, inplace=True)

print(player_stats[['WS/48']].head())
```
