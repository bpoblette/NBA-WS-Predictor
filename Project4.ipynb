```python
#Importing all the needed packages for data visualization and exploration
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


player_stats = pd.read_csv('Seasons_Stats.csv')
```

```python
sns.scatterplot(data = player_stats, x = 'PTS', y = 'WS')
```

I did several lines to see how they fit with the data. This first graph is a linear regression line

```python
sns.regplot(data=player_stats, x="PTS", y="WS", line_kws = {'color': 'red'})
```

This is quadratic regression line. Seems more fitting for the data than the linear line. Through analysis of the regressions line I have come to the conclusion that the relationship between points and win shares is quadratic in nature.

```python
sns.regplot(data=player_stats, x="PTS", y="WS", order=2, line_kws = {'color': 'red'})
```

```python
sns.regplot(data=player_stats, x="PTS", y="WS", order=3, line_kws = {'color': 'red'})
```
